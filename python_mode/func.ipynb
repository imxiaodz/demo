```python
# 定义函数
def greet(user_name): # 这里的变量 user_name 是一个形参 : 函数完成其工作所需的一项信息
	print('hello ' + user_name + '!')

greet('xdz') # 值‘xdz’是个实参 : 调用函数时传递给函数的信息.
```

```python
def bubbleSort(a):
	n = len(a)
	for i in range(n):
		for j in range(0, n-i-1):
			if a[j] > a[j+1]:
				a[j], a[j+1] = a[j+1], a[j]
	return a



arr1 = [3, 9, 4, 2, 7, 1]
a = bubbleSort(arr1)
print ("排序后的数组:", a)
```
